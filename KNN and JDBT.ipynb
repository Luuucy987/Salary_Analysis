```python
from sklearn.preprocessing import LabelEncoder
import nums_from_string
```

```python
le = LabelEncoder()
df = pd.DataFrame()
```

```python
l = []
for i in f['Grade']:
    l.append(nums_from_string.get_nums(i)[0])
```

```python
df['Gender']=f['Gender'].map({'M': 1, 'F': 0})
df['Longevity_Pay']=f['Longevity_Pay']
le.fit(f['Division'])
df['Division'] = le.transform(f['Division'])
le.fit(f['Grade'])
df['Grade'] = [i for i in l]
df['Base_Salary']=f['Base_Salary']
df['Base_Salary2']=f['Base_Salary']
```

```python
df
```

```python
mean_salary = 90000
```

```python
df['Base_Salary'] = (df['Base_Salary'] > mean_salary).astype(int)
```

```python
df
```

```python
df['Base_Salary2'] = pd.cut(df['Base_Salary2'],bins=[0,50000,100000,150000,200000,float('inf')],labels=[1,2,3,4,5])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.loc[:,'Gender':'Grade'],
                                                   df['Base_Salary'],train_size = 0.75,test_size=0.25, random_state = 1234)
print("训练数据集中共有 %d 条观测"  %X_train.shape[0])
print("测试数据集中共有 %d 条测试" %X_test.shape[0])
```

```python
y_train
```

```python
sum(y_test==1)
```

```python
# 导入K近邻模型的类
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier 

# 构件K近邻模型
kn = KNeighborsClassifier(p=5)
kn.fit(X_train,y_train.astype('int'))
print(kn)
```

```python
# 导入网格搜索函数  
from sklearn.model_selection import GridSearchCV  
  
# 选择不同的参数  
k_options = list(range(1,30))  # 生成一个从1到11的整数列表作为k值的选项  
  
parameters = {'n_neighbors': k_options}  # 创建一个字典，其中'n_neighbors'是KNeighborsClassifier的参数，k_options是对应的候选值列表  
  
# 搜索不同的K值  
# 注意：这里应该首先导入KNeighborsClassifier  
from sklearn.neighbors import KNeighborsClassifier  # 导入KNeighborsClassifier  
grid_kn = GridSearchCV(estimator=KNeighborsClassifier(), param_grid=parameters, cv=10, scoring='accuracy')  
# 创建GridSearchCV对象，其中estimator是待搜索的模型，param_grid是参数网格，cv是交叉验证的折数，scoring是评分指标  
  
grid_kn.fit(X_train, y_train)  # 使用训练数据拟合GridSearchCV对象，它会搜索最佳参数  
  
# 结果输出  
# grid_scores_ 已被弃用，在较新的scikit-learn版本中应该使用cv_results_  
# 可以使用grid_kn.cv_results_['mean_test_score']来获取每个参数组合的测试集平均得分  
# 使用grid_kn.best_params_获取最佳参数  
# 使用grid_kn.best_score_获取最佳得分  
print("Best parameters set found on development set:")  
print()  
print(grid_kn.best_params_)  
print()  
print("Grid scores on development set:")  
print()  
means = grid_kn.cv_results_['mean_test_score']  
stds = grid_kn.cv_results_['std_test_score']  
for mean, std, params in zip(means, stds, grid_kn.cv_results_['params']):  
    print("%0.3f (+/-%0.03f) for %r"  
          % (mean, std * 2, params))  
print()
```

```python
kn_pred = kn.predict(X_test)
print(pd.crosstab(kn_pred, y_test))

# 模型得分
print("模型在训练集上的准确率为%f" % kn.score(X_train, y_train))
print("模型在测试集上的准确率为%f" % kn.score(X_test, y_test))
```

```python
kn_pred
```

```python
from sklearn import metrics

# 计算ROC曲线的x轴 和 y轴数据
fpr, tpr, _ = metrics.roc_curve(y_test, kn.predict_proba(X_test)[:,1])
# 绘制ROC曲线
plt.plot(fpr, tpr, linestyle = 'solid', color ='red')
# 添加阴影
plt.stackplot(fpr, tpr, color='steelblue')
# 绘制参考线
plt.plot([0,1],[0,1],linestyle='dashed', color='black')
# 添加文本
plt.text(0.6, 0.4, 'AUC=%.3f' % metrics.auc(fpr,tpr), fontdict=dict(size =16))
plt.show()
```

```python
gbdt = GradientBoostingClassifier()
gbdt.fit(X_train, y_train) 
print(gbdt)
```

```python
from sklearn import metrics

gbdt_pred = gbdt.predict(X_test)
print(pd.crosstab(gbdt_pred, y_test))

# 模型得分
print("模型在训练集上的准确率为%f" % gbdt.score(X_train, y_train))
print("模型在测试集上的准确率为%f" % gbdt.score(X_test, y_test))

```

```python
# K 近邻模型网格搜索法
# 导入网格搜索函数
from sklearn.model_selection import GridSearchCV
# 选择不同的参数
k_options = list(range(1,12))
parameters = {'n_neighbors':k_options}
# 搜索不同的K值
grid_kn = GridSearchCV(estimator= KNeighborsClassifier(), param_grid=parameters, cv=10, scoring='accuracy')
grid_kn.fit(X_train, y_train)
# 结果输出
grid_kn.cv_results_ , grid_kn.best_params_, grid_kn.best_score_
```

```python
from sklearn import metrics

grid_kn_pred = grid_kn.predict(X_test)
print(pd.crosstab(grid_kn_pred, y_test))

# 模型得分
print("模型在训练集上的准确率为%f" % grid_kn.score(X_train, y_train))
print("模型在测试集上的准确率为%f" % grid_kn.score(X_test, y_test))

fpr, tpr, _ = metrics.roc_curve(y_test, grid_kn.predict_proba(X_test)[:,1])
plt.plot(fpr, tpr, linestyle = 'solid', color ='red')
plt.stackplot(fpr, tpr, color='steelblue')
plt.plot([0,1],[0,1],linestyle='dashed', color='black')
plt.text(0.6, 0.4, 'AUC=%.3f' % metrics.auc(fpr,tpr), fontdict=dict(size =16))
plt.show()
```

```python
# GBDT 模型的网格搜索法
# 选择不同的参数
from sklearn.model_selection import GridSearchCV
learning_rate_options = [0.01, 0.05, 0.1]
max_depth_options = [3,5,7,9]
n_estimators_options = [100, 300, 500]

parameters = {'learning_rate':learning_rate_options,
             'max_depth':max_depth_options,
             'n_estimators':n_estimators_options}

grid_gbdt = GridSearchCV(estimator= GradientBoostingClassifier(),param_grid=parameters,cv=10,scoring='accuracy')
grid_gbdt.fit(X_train, y_train)

# 结果输出
grid_gbdt.best_params_, grid_gbdt.best_score_
```

```python
from sklearn import metrics

grid_gbdt_pred = grid_gbdt.predict(X_test)
print(pd.crosstab(grid_gbdt_pred , y_test))

# 模型得分
print("模型在训练集上的准确率为%f" % grid_gbdt.score(X_train, y_train))
print("模型在测试集上的准确率为%f" % grid_gbdt.score(X_test, y_test))

fpr, tpr, _ = metrics.roc_curve(y_test, grid_gbdt.predict_proba(X_test)[:,1])
plt.plot(fpr, tpr, linestyle = 'solid', color ='red')
plt.stackplot(fpr, tpr, color='steelblue')
plt.plot([0,1],[0,1],linestyle='dashed', color='black')
plt.text(0.6, 0.4, 'AUC=%.3f' % metrics.auc(fpr,tpr), fontdict=dict(size =16))
plt.show()
```
